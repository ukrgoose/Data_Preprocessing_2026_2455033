# Simple Random Sampling (просте випадкове вибіркове дослідження)

```python
import seaborn as sns
import pandas as pd

# 모집단 (population)
tips = sns.load_dataset("tips")

# 단순 무작위 표본추출 (random sampling)
sample_random = tips.sample(n=30, random_state=42)

print(sample_random.head())
```

# Systematic Sampling (систематична вибірка)

```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")

# 간격 설정
k = 5

# 체계적 추출
sample_systematic = tips.iloc[::k]

print(sample_systematic.head())
```

#Stratified Sampling (стратифікована вибірка)


```python
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")

# 층화 추출
sample_stratified = tips.groupby('sex', group_keys=False).apply(
    lambda x: x.sample(n=15, random_state=42)
)

print(sample_stratified.head())
```
